```python
"""
author: amir kamalian
date:   22 june 2026
"""
import matplotlib.pyplot as plt
from matplotlib.cm import ScalarMappable
from matplotlib.colors import Normalize
from matplotlib.ticker import MaxNLocator
from matplotlib.colors import BoundaryNorm
import numpy as np
import pandas as pd
from scipy.optimize import curve_fit

```

```python
# data files 
CoXRF = np.loadtxt("../../LL3S/LL3S/XRF/LL3SCo.txt", delimiter = ';')      
GdXRF = np.loadtxt("../../LL3S/LL3S/XRF/LL3SGd.txt", delimiter = ';')


# molar masses of Gd and Co
MGd = 157.25
MCo = 28.01

pGd = 7.900
pCo = 8.900


# discard values above 350 to remove outter edge scan values 
mask = (0 < GdXRF) & (GdXRF < 350)
GdXRF = mask*GdXRF
CoXRF = mask*CoXRF


# creating an xy coordinate grid to plot against based on size of data array
posX = np.linspace(-5.8, 5.8, CoXRF.shape[1]) 
posY = np.linspace(5.8, -5.8, CoXRF.shape[0])
X, Y = np.meshgrid(posX, posY, indexing="xy") # (columns, rows)


# ====================================
# PLOTTING
# ====================================
fig, axes = plt.subplots(1, 2, figsize=(16, 6), gridspec_kw={"width_ratios" : [1,1]})

# for color bars
normCo = Normalize(
    vmin=CoXRF.min(),
    vmax=CoXRF.max()
)

co_sm = ScalarMappable(norm=normCo, cmap=plt.cm.GnBu)
co_sm.set_array([])

normGd = Normalize(
    vmin=GdXRF.min(),
    vmax=GdXRF.max()
)
gd_sm = ScalarMappable(norm=normGd, cmap=plt.cm.PuRd)
gd_sm.set_array([])


# Cobalt
axes[0].pcolormesh(X, Y, CoXRF, cmap='GnBu')   ###  
axes[0].set_ylabel('Distance (cm)')
axes[0].set_xlabel('Distance (cm)')
axes[0].grid(which="major", color="#8d8f91", linestyle=":")

cbar = fig.colorbar(co_sm, ax=axes[0], pad=0.02, aspect=40)
cbar.set_label('Intensity of Co-kα Peaks (Counts)')

# Gadolinium
axes[1].pcolormesh(X, Y, GdXRF, cmap='PuRd')   ###  
axes[1].set_ylabel('Distance (cm)')
axes[1].set_xlabel('Distance (cm)')
axes[1].grid(which="major", color="#8d8f91", linestyle=":")

cbar = fig.colorbar(gd_sm, ax=axes[1], pad=0.02, aspect=40)
cbar.set_label('Intensity of Gd-Lα Peaks (Counts)')

# plt.savefig("C:\\Users\\Lily1\\LL3S\\Images\\CoGdXRF.pdf", format = "pdf", bbox_inches='tight')
plt.show()
```

```python
"""
COPIED DIRECTLY FROM LILY CODE: needed to calculate the composition map of sample; based on 
samples thickenss estimate 80nm for Gd.
"""

cmap = plt.colormaps['Reds']

DiaD = np.linspace(0, np.sqrt(288), len(CoXRF))
CoIoft = np.diagonal(np.fliplr(CoXRF))
GdIoft = np.diagonal(GdXRF)

(CoDIACURVE, CoCURVE) = (DiaD[np.logical_and(5<DiaD, DiaD<13)], CoIoft[np.logical_and(5<DiaD, DiaD<13)])
(GdDIACURVE, GdCURVE) = (DiaD[np.logical_and(5.5<DiaD, DiaD<13.5)], GdIoft[np.logical_and(5.5<DiaD, DiaD<13.5)])

def linear(x, m, b):
    return m * x + b

#Curve fit Co
poptCo, pcovCo = curve_fit(linear, CoDIACURVE, CoCURVE)
mCo, p1o = poptCo
mCoU, p1oU = np.sqrt(np.diag(np.abs(pcovCo)))


##Curve fit Gd
poptGd, pcovGd = curve_fit(linear, GdDIACURVE, GdCURVE)
mGd, bGd = poptGd
mGdU, bGdU = np.sqrt(np.diag(np.abs(pcovGd)))

print(mGdU / mGd, bGd)
print(mCoU / mCo, p1o)

mtCo = 110 / 8  ##nm / cm (curves based on mike johnson
mtGd = 10       ##nm / cm

MGd = 157.25
MCo = 28.01

pGd = 7.900
pCo = 8.900

ImaxCo = CoXRF.max() 
##m = counts / cm = (scale factor) * thickness / cm 
##thickness = counts / scale factor
##tmax = imax / scale factor %%%%%%%%%%%% scale factor = Imax / 110nm#

sCo = ImaxCo / 110 
sCoTrue = np.abs(mCo) / np.abs(mtCo)
tCoLL3S = CoXRF / sCo         ###im allowed to do this as both linearly relate to distance 
stCoLL3S = CoXRF / sCoTrue
##slope * distance from 0 = I
ImaxGd = 310

sGd = ImaxGd / 80                    ##estimated sputter deposition of gadolinium  
sGdTrue = np.abs(mGd) / np.abs(mtGd)
tGdLL3S = GdXRF / sGd                ###im allowed to do this as both linearly relate to distance 
stGdLL3S = GdXRF / sGdTrue 

tLL3S = tGdLL3S + tCoLL3S + 0.01 - (0.01 * mask)

molCo = pCo * tCoLL3S / MCo
molGd = pGd * tGdLL3S / MGd

smolCo = pCo * stCoLL3S / MCo
smolGd = pGd * stGdLL3S / MGd

"""
xLL3S corresponds to the coordinate; these are the composition values 
"""
xLL3S = molCo / (molGd + molCo + 0.01)
sxLL3S = smolCo / (smolGd + smolCo + 0.01)

print(f"    (columns, rows)")
print(f"xy: ({len(posX)}, {len(posY)})")
print(f"    (rows, columns)")
print(f"xLL3S: {xLL3S.shape}")
print(f"tLL3S: {tLL3S.shape}")


##graphing

XTlevelsx = MaxNLocator(nbins=31).tick_values(xLL3S.min(), xLL3S.max())
XTnormx = BoundaryNorm(XTlevelsx, ncolors=cmap.N, clip=True)

XTlevelst = MaxNLocator(nbins=31).tick_values(tLL3S.min(), tLL3S.max())
XTnormt = BoundaryNorm(XTlevelst, ncolors=cmap.N, clip=True)

fig, XT = plt.subplots(1, 2, figsize = (13, 5), constrained_layout = True)

CoCXT = XT[0].pcolormesh(X, Y, xLL3S, cmap = 'Greens', norm = XTnormx)   ### 
XT[0].set_xlabel('Distance (cm) x')
XT[0].set_ylabel('Distance (cm) y')
XT[0].set_box_aspect(1)
XT[0].grid(which="major", color="#8d8f91", linestyle=":")

cbar = fig.colorbar(CoCXT, ax=XT[0])
cbar.set_label('Composition of Film')

GdCXT = XT[1].pcolormesh(X, Y, tLL3S, cmap = 'Purples', norm = XTnormt)   ### 
XT[1].set_ylabel('Distance (cm) y')
XT[1].set_xlabel('Distance (cm) x')
XT[1].set_box_aspect(1)
XT[1].grid(which="major", color="#8d8f91", linestyle=":")
XT[1].scatter(X[0][28], Y[6][0]) 
cbar = fig.colorbar(GdCXT , ax=XT[1])
cbar.set_label('Thickness of Film (nm)')
# plt.savefig("C:\\Users\\Lily1\\LL3S\\Images\\BootlegXT.pdf", format = "pdf", bbox_inches='tight')
plt.show()


print(X[0])
print(Y[:,0])

```

```python
"""
WRITING A FUNCTION TO RETURN COMPOSITION BASED ON INPUT COORDINATE USING INTERPOLATION.

scipy.interpolate.RegularGridInterpolator: used for multivariate interpolation on n-dimensional grid. Spacing between orthogonal 
        does not have to be uniform. Supports multiple methods; here we will be using linear interpolation. 

"""
from scipy.interpolate import RegularGridInterpolator
%matplotlib widget

print(f"    (columns, rows)")
print(f"xy: ({len(posX)}, {len(posY)})")
print(f"    (rows, columns)")
print(f"xLL3S: {xLL3S.shape}")
print(f"tLL3S: {tLL3S.shape}")

# interpolation step
composition_inter = RegularGridInterpolator(
    (posY, posX),
    xLL3S,
    method="linear",
    bounds_error=False
)

# =================================================
# FUNCTION RETURNS THE COMPOSITION AT XY COORDINATE
# =================================================
def composition(x, y):
    return float(composition_inter((y, x)))

# return a tuple of the composition of the alloy
def alloy_composition(x, y):
    co = float(composition_inter((y, x)))
    gd = 1 - co
    return gd, co

"""
# converting the interpolated points to compatible array for plotting
mask_comp_data = np.where(xLL3S == 0, np.nan, xLL3S) 

print(comp_data.shape)
print(mask_comp_data.shape)

# 3D plot of composition 

fig, axes = plt.subplots(figsize=(13,8), subplot_kw={"projection" : "3d"})

comp_surface = axes.plot_surface(X, Y, mask_comp_data, cmap="berlin", edgecolor="none", linewidth=0, rstride=1, cstride=1)
axes.set_title("Co composition of thin film")
axes.set_xlabel("Distance (cm) x")
axes.set_ylabel("Distance (cm) y")
axes.view_init(elev=10, azim=55)

cbar = fig.colorbar(comp_surface, ax=axes, pad=0.01, shrink=0.7, aspect=30)
cbar.set_label(r"Composition x ($\text{Gd}_{1-x}\text{Co}_{x}$)")


plt.show()
"""
```

```python
# same plot as above but with finer detailing 

x_fine = np.linspace(posX.min(), posX.max(), 3*len(posX))
y_fine = np.linspace(posY.min(), posY.max(), 3*len(posY))

x, y = np.meshgrid(x_fine, y_fine)

# points to evaluate composition
points = np.column_stack([
    y.ravel(),
    x.ravel()
])

composition_fine = composition_inter(points).reshape(x.shape)
mask_comp_fine = np.where(composition_fine == 0, np.nan, composition_fine)

print(x.shape)
print(y.shape)
print(composition_fine.shape)

# 3D plot of composition 

fig, axes = plt.subplots(figsize=(13,8), subplot_kw={"projection" : "3d"})

comp_surface = axes.plot_surface(x, y, mask_comp_fine, cmap="berlin", edgecolor="none", linewidth=0, rstride=1, cstride=1)
axes.set_title("Co composition of thin film")
axes.set_xlabel("Distance (cm) x")
axes.set_ylabel("Distance (cm) y")
axes.view_init(elev=10, azim=55)

cbar = fig.colorbar(comp_surface, ax=axes, pad=0.01, shrink=0.7, aspect=30)
cbar.set_label(r"Composition x ($\text{Gd}_{1-x}\text{Co}_{x}$)")

plt.show()
```
